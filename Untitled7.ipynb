```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report,confusion_matrix
%matplotlib inline
```

```python
dataset = pd.read_csv('score.csv')
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
x = dataset[['mtk', 'ips', 'ipa','sbk','komputer']]
y = dataset['Gender']
```

```python
scalerer = StandardScaler()
scalerer.fit(x)
scaled_feature = scalerer.transform(x)
```

```python
X_train, X_test, y_train, y_test = train_test_split(scaled_feature,dataset['Gender'], test_size=0.20)
```

```python
print("total siswa laki laki:", X_train.shape[0])
print("total siswa perempuan:", X_test.shape[0])
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Inisialisasi list untuk menyimpan hasil
k_values = []
accuracy_values = []

# Melakukan iterasi untuk setiap nilai K yang diinginkan
for k in range(1, 40):
    # Membuat objek KNN dengan n_neighbors=k
    knn = KNeighborsClassifier(n_neighbors=k)

    # Melatih model KNN menggunakan data latih
    knn.fit(X_train, y_train)

    # Melakukan prediksi pada data uji
    pred = knn.predict(X_test)

    # Menghitung akurasi
    accuracy = accuracy_score(y_test, pred)

    # Menyimpan nilai K dan akurasi
    k_values.append(k)
    accuracy_values.append(accuracy)

# Membuat DataFrame dari hasil
df = pd.DataFrame({'K': k_values, 'Accuracy': accuracy_values})

# Menampilkan tabel
print(df)
```

```python
error_rate = []

for i in range(1,39):

    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))
```

```python
plt.figure(figsize=(10, 6))
plt.plot(range(1, 39), error_rate, color='k', linestyle='dashed', marker='o', markersize=5)
plt.title('Error Rate vs. Nilai K')
plt.xlabel('K')
plt.ylabel('Error Rate')
plt.xticks(range(1, 39))
plt.show()
```
